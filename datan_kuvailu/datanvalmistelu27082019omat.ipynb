```python
df["Taajama-aste"] = df["Taajama-aste, %, 2017"]
df["Ruotsinkieliset"] = df["Ruotsinkielisten osuus väestöstä, %, 2018"]
df_ruotsi = df[["Kunta","Ruotsinkieliset", "Taajama-aste"]][df["Ruotsinkieliset"]>=50]
df_ruotsi = df_ruotsi.sort_values(by="Ruotsinkieliset", ascending=False)
df_ruotsi.head(30).sort_values(by="Taajama-aste", ascending=False)
```
