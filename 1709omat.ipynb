```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.style.use("fivethirtyeight")
```

```python
happiness_alku = pd.read_excel("http://taanila.fi/happy.xlsx")
happiness = happiness_alku[happiness_alku.Year == 2017].dropna() #otetaan mukaan vain vuoden 2017 tiedot
happiness
```

```python
happiness.drop("Year", axis=1).describe().style.format("{:.1f}")
```

```python


happiness[["Happiness Score", "Economy (GDP per Capita)", "Health (Life Expectancy)","Freedom", "Trust (Government Corruption)"
        ]].corr().style.format("{:.2f}")
```

```python
happiness.shape
```

```python
bins=[0,4.5,5.3,6.2,8]
sns.distplot(a=happiness['Happiness Score'].dropna(), bins=bins, hist=True, kde=False, rug=False)
```

```python
kuvio1= happiness.plot.scatter("Economy (GDP per Capita)", "Happiness Score")
kuvio1.get_figure().suptitle("")
kuvio1.set_title("The relationship between GDP per Capita and Happiness Score")
```
